## Creating subagents

```python
from langchain.tools import tool

@tool   # Define tool 1
def square_root(x: float) -> float:
    """Calculate the square root of a number"""
    return x ** 0.5

@tool   # Define tool 2
def square(x: float) -> float:
    """Calculate the square of a number"""
    return x ** 2
```

```python
from langchain.agents import create_agent

# create subagents as tools

subagent_1 = create_agent(
    model='gpt-5-nano',
    tools=[square_root]
)

subagent_2 = create_agent(
    model='gpt-5-nano',
    tools=[square]
)
```

## Calling subagents

```python
from langchain.messages imoprt HumanMessage

@tool
def call_subagent_1(x: float) -> float:
    """ Call subagent 1 in order to calculate the square root of a number"""
    response = subagent_1.invoke({"messages": [HumanMessage(content = f"Calculate the square root of {x}")]})
    return float(response.content)

@tool
def call_subagent_2(x: float) -> float:
    """Call subagent 2 in or der to calculate the square of a number"""
    response = subagent_2.invoke({"messages": [HumanMessage(content = f"Calculate the square of {x}")]})
    return float(response.content)

# Create the main agent
main_agent = create_agent(
    model = "gpt-5-nano",
    tools = [call_subagent_1, call_subagent_2],
    system_prompt = "You are a helpful agent who can call subagents to calculate the square root and square of a number."
)
```

## Test

```python
question = "What is the square root of 9?"

response = main_agent.invoke({"messages": [HumanMessage(content = question)]})
print(response["messages"][-1].content)
```
